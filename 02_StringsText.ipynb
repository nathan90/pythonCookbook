```python
file_path = 'part-00008-33d369a7-eac6-4a4b-9771-831d69e4aef6-c000 (2).json'
```

```python
import json
jasonarray = []
with open(file_path) as f:
    for line in f:
        j_content = json.loads(line)
#         print(j_content, '\n')
        jasonarray.append(j_content['value'].copy())
```

```python
import pandas as pd
x =pd.json_normalize(jasonarray)#, record_path='value')
```
